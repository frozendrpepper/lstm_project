## 2) LSTM Basics and some examples

Below diagram shows the basic infrastructure of a LSTM, arguably the most popular node used in RNN. Other options include SimpleRNN and GRU but, as far as I'm concerned, LSTM outperforms both of them in every aspect. I'll focus on LSTM here

Besides all the inner working, we need to notice that for every input vector **Xt** into a LSTM node, we get a memory signal **Ct** and output **ht** both of which can either be passed to next LSTM node or fed into another network for further analysis (i.e softmax layer for classification of results from that particular block).
