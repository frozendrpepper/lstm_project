### 2-1) Simple RNN example to reproduce a song

#### 2-1-1) Basics of LSTM layer in Keras

LSTM(cell_number, input_dim, input_length, return_sequence, stateful)

* cell_number = equivalent to number of nodes in each layer of Neural Network, more cell numbers result in more complex kernel at the expense of computational time.
* input_dim = equivalent to number of input dimension in Neural Network. If each of you sequence data is only single dimensional, insert 1 otherwise to the number corresponding to the dimensions of each sequence data.
* input_length = indicates how many data points of input_dim will be entered in one sequence. For instance, if individual data is single letter the input_dim = 1. But if the model uses 4 sequences of letters input_length = 4.
* return_sequence = indicates whether to return single output from the LSTM layer or full sequence output. For many to one, use the default value which is False, and for many to many models set it equal to True.
* stateful = Default false. If set to true, the final state of one batch is used as initial state for the next batch.

#### 2-1-2) Data

Our sequence data consists of music pitch + note length for a song. For instance, c8 means note corresponding to C and 1/8th note in length

```python
seq = ['g8', 'e8', 'e4', 'f8', 'd8', 'd4', 'c8', 'd8', 'e8', 'f8', 'g8', 'g8', 'g4', 
       'g8', 'e8', 'e8', 'e8', 'f8', 'd8', 'd4', 'c8', 'e8', 'g8', 'g8', 'e8', 'e8', 'e4',
       'd8', 'd8', 'd8', 'd8', 'd8', 'e8', 'f4', 'e8', 'e8', 'e8', 'e8', 'e8', 'f8', 'g4',
       'g8', 'e8', 'e4', 'f8', 'd8', 'd4', 'c8', 'e8', 'g8', 'g8', 'e8', 'e8', 'e4']
```

```python
#Dictionaries used to convert each data point to numeric values and vice versa
code2idx = {'c4':0, 'd4':1, 'e4':2, 'f4':3, 'g4':4, 'a4':5, 'b4':6,
            'c8':7, 'd8':8, 'e8':9, 'f8':10, 'g8':11, 'a8':12, 'b8':13}

idx2code = {0:'c4', 1:'d4', 2:'e4', 3:'f4', 4:'g4', 5:'a4', 6:'b4',
            7:'c8', 8:'d8', 9:'e8', 10:'f8', 11:'g8', 12:'a8', 13:'b8'}
```

```python
#Helper function used to dissect and convert data into form that Keras can accept
def seq2dataset(seq, window_size):
    dataset = []
    for i in range(len(seq) - window_size):
        #First dissect the sequence into window_size + 1 for element
        #The last +1 element would function as the label for that particular sequence
        subset = seq[i:(i + window_size + 1)]
        dataset.append([code2idx[item] for item in subset])
    return np.array(dataset)
```

```python
dataset = seq2dataset(seq, window_size = 4)
print(dataset)
```

In the above datset, first 4 numbers in each array would correspond to the input to our LSTM which uses input_dim = 4 and the last digit in each array corresonds to the label or the next note that the LSTM model is trying to predict.

##### Divide the data set into x and y

```python
x_train = dataset[:, 0:4]
y_train = dataset[:, 4]
print(x_train)
```

```python
#This value is used to normalize the x_train 
max_index_value = 13
```

```python
x_train = x_train / float(max_index_value)
```

```python
print(x_train)
```

#### Note about input dimensions for LSTM

LSTM model in Keras takes inputs with following dimensions:

**(number_samples, timeseries, input_dim)**

* number_samples = Number of datapoints
* timeseries = equivalent input_length in LSTM model
* input_dim = dimension of each individual input x

```python
dataset.shape
```

We have 50 data samples

```python
x_train = np.reshape(x_train, (50, 4, 1))
```

```python
print(x_train)
```

```python
#One hot encode the label
y_train = np_utils.to_categorical(y_train)
print(y_train)

#Make sure one hot encoding is done properly
one_hot_vec_size = y_train.shape[1]
print("One hot encoded vector size is: ",one_hot_vec_size)
```

So out of 13 possible notes in the song, only 12 ended up being in the label

#### 2-1-2) Model building

We will build a LSTM model by stacking up layers in the Sequential layer. The final single output from LSTM is fed into a dense layer with softmax activation and the integer value corresponding to the number of unique labels for multi-class classification.

Also, we will use LSTM with 128 units and input_shape = (input_length, input_dim)

```python
model = Sequential()
model.add(LSTM(128, input_shape = (4, 1)))
model.add(Dense(one_hot_vec_size, activation = 'softmax'))
```

#### 2-1-3) Model compile and fit

* loss is categorical crossentropy which is basically equivalent to log loss for multi class classification problem which this problem is. 
* Adam optimizer is a good default optimizer to use for faster and robust convergence
* The optimizer will use accuracy as its metric to find the optimal solution

```python
model.compile(loss = 'categorical_crossentropy', optimizer = 'adam', metrics = ['accuracy'])

#LossHistory class we defined early to keep track of losses at each epoch
history = History()

model.fit(x_train, y_train, epochs = 2000, batch_size = 14, verbose = 2, callbacks=[history])
```

#### 2-1-4) Visualizing loss and accuracy

```python
plt.plot(history.history['loss'])
plt.ylabel('loss')
plt.xlabel('epoch')
plt.title('LSTM Loss History')
plt.show()
```

```python
plt.plot(history.history['acc'])
plt.ylabel('acc')
plt.xlabel('epoch')
plt.title('LSTM Accuracy History')
plt.show()
```

```python
#Evaluating the model
scores = model.evaluate(x_train, y_train)
print(model.metrics_names[1], scores[1] * 100)
```
